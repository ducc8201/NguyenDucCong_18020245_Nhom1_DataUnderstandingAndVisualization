```python
labels = 'Thể thao', 'Bóng đá', 'Đời sống Showbiz', 'Pháp luật', 'Giải trí', 'Tin tức trong ngày', 'Du lịch', 'Thế giới', 'Khác'
labels1 = 'TT', 'BĐ', 'DSS', 'PL', 'GT', 'TTTN', 'DL', 'TG', 'K'
sizes = [1416, 1067, 850, 628, 514, 511, 505, 366, 3835]
plt.pie(sizes,labels=labels,autopct='%1.1f%%', startangle=15, shadow = True)
plt.title('Phân loại')
plt.axis('equal')
plt.show()
```

```python
plt.xlabel("Thể loại")
plt.ylabel("Số bài đăng")
plt.suptitle("Số bài đăng theo phân loại")

for bar in plt.bar(labels1, sizes):
    plt.text(bar.get_x() + .4,
            bar.get_height(),
            bar.get_height(),
            ha = 'center',
            va = 'bottom',
            color = 'black')
```
